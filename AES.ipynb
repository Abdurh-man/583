ffAdd() - adds two finite fields (see Section 4.1)

```python
def ffAdd(x1, x2):
#     a = format(x1, '0>8b')
#     b = format(x2, '0>8b')
    a = [int(i) for i in "{0:08b}".format(x1)]
    b = [int(i) for i in "{0:08b}".format(x2)]
    result = 0b00000000
    addition = []
    for i in range(len(a)):
        bit =  (a[i]^b[i])
        result |= bit << 7 - i
    return result
```

xtime() - multiplies a finite field by x (see Section 4.2.1)

```python
#according to the documentation xtime(b) returns xb(x) mod x8+x4+x3+x+1 
def xtime(x):
    temp = x
    x = (x << 1)&(0xff)
    #if leading bit is one 
    if temp & 0x80:  
        x ^= 0x1b 
        #beacuse, we use x^4 + x^3 + x + 1 which is 11011 

    return x
```

ffMultiply() - uses xtime to multiply any finite field by any other finite field. (see Section 4.2.1)

```python
#0x57 0101 0111
#0x13 0001 0011 
def ffMultiply(a, b):
    ans = 0
    while(a != 0 and b != 0):
        if (b & 0x01):
            ans ^= a

        a = xtime(a)
        b >>= 1

    return ans
```

subWord() - takes a four-byte input word and substitutes each byte in that word with its appropriate value from the S-Box. The S-box is provided (see Section 5.1.1).

```python
def subWord(word):
    x1,y1 = (word>>28)&0x0f, (word>>24)&0x0f
    x2,y2 = (word>>20)&0x0f, (word>>16)&0x0f
    x3,y3 = (word>>12)&0x0f, (word>>8)&0x0f
    x4,y4 = (word>>4)&0x0f, (word>>0)&0x0f

    sword = 0x00000000
    s1 = (Sbox[x1][y1])
    s2 = (Sbox[x2][y2])
    s3 = (Sbox[x3][y3])
    s4 = (Sbox[x4][y4])

    sword = (sword << 8) + s1
    sword = (sword << 8) + s2
    sword = (sword << 8) + s3
    sword = (sword << 8) + s4
    return sword

# def subWord(word):
#     b1, b2, b3, b4 = ((word & 0xff000000) >> 24), ((word & 0xff0000) >> 16), ((word & 0xff00) >> 8), (word & 0xff)
#     state = [b1, b2, b3, b4]
#     ret_word = 0
#     for byte in state:
#         x, y = ((byte & 0xf0) >> 4), (byte & 0xf)
#         ret_word = (ret_word << 8) + sBox[x][y]
#     return ret_word
```

rotWord() - performs a cyclic permutation on its input word.

```python
def rotWord(key):
    low = key >> 24
    key = key & 0x00ffffff
    key = key << 8 | low
    return key
```

```python
def keyExpansion(key,w, Nk=4, Nr=10):
    temp = ""
    i = 0
    while (i < Nk):
        #          24          16          8          0
        w[i] = (key[4 * i] << 24) | (key[4 * i + 1] << 16) | (key[4 * i + 2] << 8) | (key[4 * i + 3])
        i = i+1
        
    i = Nk
    while (i < len(w)):
        temp = w[i-1]

        if (i%Nk == 0):
            temp = subWord(rotWord(temp)) ^ Rcon[int(i/Nk)]
        
        elif(Nk > 6 and i%Nk == 4):
            temp = subWord(temp)

        w[i] = w[i-Nk] ^ temp
        i = i + 1
```

shiftRows() - This transformation performs a circular shift on each row in the State (see Section 5.1.2)

```python
def shiftRows(state):
    key = 0x00000000
    for i, ivalues in enumerate(state): 
        key = 0x00000000
        for j, jvalues in enumerate(ivalues):
            val = 0x00000000
            val |= jvalues

            key |= val << (24-(j*8))
            
        for j in range(i):
            key = rotWord(key)
            key = key
        
        state[i][3] = ((key>>0)&0x000000ff)
        state[i][2] = ((key>>8)&0x000000ff)
        state[i][1] = ((key>>16)&0x000000ff)
        state[i][0] = ((key>>24)&0x000000ff)
```

mixColumns() - This transformation treats each column in state as a four-term polynomial. This polynomial is multiplied (modulo another polynomial) by a fixed polynomial with coefficients (see Sections 4.3 and 5.1.3).

```python
def mixColumns(state):
    #look at aes documentation page 18
    for i in range(4):
        a = state[0][i]
        b = state[1][i]
        c = state[2][i]
        d = state[3][i]
        s0 = ffMultiply(a, 0x02) ^ ffMultiply(b, 0x03) ^ ffMultiply(c, 0x01) ^ ffMultiply(d, 0x01)
        s1 = ffMultiply(a, 0x01) ^ ffMultiply(b, 0x02) ^ ffMultiply(c, 0x03) ^ ffMultiply(d, 0x01)
        s2 = ffMultiply(a, 0x01) ^ ffMultiply(b, 0x01) ^ ffMultiply(c, 0x02) ^ ffMultiply(d, 0x03)
        s3 = ffMultiply(a, 0x03) ^ ffMultiply(b, 0x01) ^ ffMultiply(c, 0x01) ^ ffMultiply(d, 0x02)

        state[0][i] = (s0)    
        state[1][i] = (s1)    
        state[2][i] = (s2)    
        state[3][i] = (s3)    
```

```python
def to_string(state):
    sword = 0
    #     print(state)
    for i, ivalue in enumerate(state):
        for j, jvalue in enumerate(ivalue):
            sword = (sword << 8) + jvalue

    sword = hex(sword)
    return sword[2:]
```

```python
def decrypt(data_in, data_out,w, Nk = 4, Nr = 10, Nb = 4):
    state = []
    state = data_in

    addRoundKey(state,w)
    for round in range(Nr+1,0,-1):
        print(f'round[{Nr-round+1:2}].istart   ',to_string(state))
        invSubByte(state)
        print(f'round[{Nr-round+1:2}].is_row   ',to_string(state))
        invShiftRows(state)
        print(f'round[{Nr-round+1:2}].is_box   ',to_string(state))
        invMixColumns(state)
        print(f'round[{Nr-round+1:2}].ik_sch   ',to_string(state))
        addRoundKey(state, w)
        print(f'round[{Nr-round+1:2}].ik_add   ',to_string(state))

    subByte(state)
    shiftRows(state)
    addRoundKey(state, w)
    print(f'round[{Nr:2}].output   ',to_string(state))
```
